```python
#HEATMAP
plt.figure(figsize = (10,6))
annot_kws = {"fontsize":10,
            "fontstyle":"italic",
            'color':'k',
            'alpha':0.9,
            'verticalalignment':'center',
            "backgroundcolor":'w'}
cbar_kws = {"orientation":"vertical", 
            "shrink":1,
            'extend':'min',
            'extendfrac':0.1,
            'ticks':np.arange(0,22),
            "drawedges":True }
sns.heatmap(data.corr(), vmax=1.5,cmap = "gist_ncar_r",annot = True,annot_kws=annot_kws,linewidth=2,linecolor='k',cbar_kws=cbar_kws)
```
